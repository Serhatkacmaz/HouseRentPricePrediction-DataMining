```python
import time
import pandas as pd
import random

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
```

```python
href_list = []
ex_href_list = []
all_Home_List = []
count = 0

def listToString(s):
    str1 = ""
    return (str1.join(s))

def strRegex(str):
    return str[0].replace("M2","").replace("m2","").replace(". Kat","").replace("TL","").replace("Yaşında","").replace("Katlı","").strip()
```

```python
# both
paging_Off_set = 40
random_num = 1

#link topla
s = Service("C:\DataMining\chromedriver_win32\chromedriver.exe")
driver = webdriver.Chrome(service=s)
driver.maximize_window()

while paging_Off_set <= 980:
    
    href_list = []
    time.sleep(random.randint(3, 9))
    
    try:
        driver.get("https://www.sahibinden.com/kiralik-daire/istanbul?pagingOffset=" + str(paging_Off_set) + "&sorting=price_asc&price_min=4000&a20=38473&a20=38470&a20=38474&a20=38471&query_text_mf=kiral%C4%B1k&price_max=30000&query_text=kiral%C4%B1k")
    except:
        print("page exception")
        continue
        
    #cıkan reklam engel
    #driver.execute_script("document.getElementsByClassName('feature-discovery__big-circle')[0].style.display='none';")

    # ilan linki
    liste_item = driver.find_elements(by=By.CLASS_NAME, value="classifiedTitle")

    for i in range(len(liste_item)):
        href_list.append(liste_item[i].get_attribute('href'))
        
    # linkler bilgi csv
    for href_num in href_list:
        count = count + 1
        time.sleep(random.randint(0,random_num))
        try:
            
            driver.get(href_num)

            # ilan içeriği
            pull_elements = driver.find_elements(by = By.CLASS_NAME, value="classifiedInfoList")  # Teknik Özellik
            pull_price = driver.find_element(by = By.XPATH, value = "//*[@id='favoriteClassifiedPrice']")  # fiyat
            pull_price = pull_price.get_attribute('value')

            ilce = driver.find_element(by=By.CSS_SELECTOR, value="#classifiedDetail > div > div.classifiedDetailContent > div.classifiedInfo > h2 > a:nth-child(3)")  


            # özellikler
            list_elements = []
            for element in pull_elements:
                list_elements.append(element.text)


            # düzenle
            list_elements_str = listToString(list_elements)
            elements_splitter = list_elements_str.split("\n")
            df = pd.DataFrame(elements_splitter)

            df = df.iloc[6:30]
            df = df.reset_index()  # değişiklik index kayma fix
            df.drop("index", axis=1, inplace=True)  # oluşan index kaldır
            
            # özellikler
            df_list = df.values.tolist()
            list_home_feature = []
            for i in range(len(df_list)):
                if (i % 2 == 1):
                    list_home_feature.append(strRegex(df_list[i]))

            list_home_feature.append(ilce.text.strip())
            list_home_feature.append(pull_price.strip())  # kira val ekle

            all_Home_List.append(list_home_feature)
            paging_Off_set = paging_Off_set + 20
            print("-- " + str(count))
            
            random_num = random_num + 1
            
            if random_num == 10:
                random_num = 1
            
        except:
            print("exception")
            ex_href_list.append(href_num)

#save to csv
df_home_features = pd.DataFrame(all_Home_List)

df_home_features.to_csv("sahibinden.csv",
                        encoding="utf-8", index=False)

driver.quit()
print('-- ok')
```
